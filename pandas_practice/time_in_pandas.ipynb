```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

data = pd.DataFrame({
    "student_id": range(1, 301),
    "course_name": np.random.choice(
        ["Python", "Data Science", "ML", "Web Dev"], 300
    ),
    "progress_percent": np.random.randint(30, 101, 300),
    "completion_days": np.random.randint(10, 120, 300),
    "rating": np.random.randint(1, 6, 300)
})

data.head()
```

```python
course_summary = data.groupby("course_name").mean()
course_summary

```

```python
poor_courses = course_summary[course_summary["progress_percent"] < 70]

poor_courses
```
