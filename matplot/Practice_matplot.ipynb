```python
df['experience']=[1,2,3,1,1.5,2.5,1,1,2.5,3]
df.head()
```

```python
plt.scatter(df['Age'],df['salary'],s=df['experience']*200,color='skyblue',edgecolors='black')
plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("salary")
plt.grid()
```

```python
color={'hr':'yellow','it':'blue','finance':'orange'}
for dept,color in color.items():
    df_dep=df[df['dept']==dept]
    plt.scatter(df_dep["Age"],df_dep["salary"],c=color,label=dept)
plt.legend()
```

object oriented api

```python
sorted_age
```

```python
fig,axs=plt.subplots(1,3,figsize=(15,5))
#line plot
axs[0].plot(sorted_age["Age"],df["salary"],color='red',marker="*",linewidth="2",markersize=2)
axs[0].grid()
axs[0].set_title("line plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

#histogram
axs[1].hist(df["salary"],color='skyblue',bins=5)
axs[1].set_title("line plot")
axs[1].set_xlabel("salary")
axs[1].set_ylabel("Frequency")

#boxplot
axs[2].boxplot(df["salary"])
axs[2].set_title("box plot")
axs[2].set_xlabel("Salary")
plt.savefig("obj_mplots.png")
plt.show()
```
